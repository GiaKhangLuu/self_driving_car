```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt

from model import load_data, build_model
from utils import batch_generator
```

```python
weight_file_path = 'models/model-004.h5'

X_train, X_valid, y_train, y_valid = load_data()

model = build_model()
model.load_weights(weight_file_path)
```

```python
# Re-train model
epochs = 10
batch_size = 32
steps_per_epoch = len(X_train) // batch_size
validation_steps = len(X_valid) // batch_size

checkpoint = ModelCheckpoint('models/model-{epoch:03d}.h5',
                             monitor='val_loss',
                             verbose=1,
                             save_best_only=True)

model.compile(optimizer='adam', loss='mean_squared_error')

history = model.fit(batch_generator('dataset', X_train, y_train, batch_size, True),
                    steps_per_epoch=steps_per_epoch,
                    epochs=epochs,
                    validation_data=batch_generator('dataset', X_valid, y_valid, batch_size, True),
                    validation_steps=validation_steps,
                    callbacks=[checkpoint],
                    verbose=1)
```

```python
# Plot learning curve
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.title('Model loss')
plt.xticks(range(epochs))
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
